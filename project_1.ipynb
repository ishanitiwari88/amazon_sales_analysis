```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Reading data from csv file:
data = pd.read_csv('amazon.csv')
data.head()
```

```python
#convert the 'rating' column to numeric, coercing the errors
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

#fill the empty cells with 0
data["rating"].fillna(0)

#applying a filter which filters the ratings that are greater that 4:
filt = data['rating']>4
data[filt].head()
```

```python
# # Mapping each unique combination of categories as a whole to a unique number

# # Extracting unique combinations from the 'category' column
# unique_combinations = data['category'].unique()

# # Creating a mapping for all unique combinations to numbers
# combination_mapping = {combination: i for i, combination in enumerate(unique_combinations)}

# # Applying the mapping to the 'category' column
# data['category'] = data['category'].map(combination_mapping)

# # Displaying the last few rows after the transformation
# data['category'].sort_values().tail()
# Extracting unique combinations from the 'category' column
unique_combinations = data['category'].unique()

# Creating a mapping for all unique combinations to numbers
combination_mapping = {combination: i for i, combination in enumerate(unique_combinations)}

# Create a reverse mapping (number to category name)
reverse_mapping = {i: combination for combination, i in combination_mapping.items()}

# Applying the mapping to create a new column 'category_num'
data['category_num'] = data['category'].map(combination_mapping)

# Displaying the last few rows after the transformation
print(data[['category', 'category_num']].sort_values('category_num').tail())

# Printing the first 10 categories which have occurred the most
count = data['category_num'].value_counts()[0:10]
count_with_names = count.rename(index=reverse_mapping)
```

```python
#now we have seen that we have 210 categories in total.
#printing the first 10 categories which has occured the most
count = data['category'].value_counts()[0:10]
print("\nTop 10 most frequent categories:")
count
```

```python

def truncate_string(string, max_length=20):
    return string[:max_length] + '...' if len(string) > max_length else string


truncated_count = count.copy()
truncated_count.index = [truncate_string(name) for name in count.index]

plt.figure(figsize=(12, 6))
truncated_count.plot(kind='bar', color='green')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Top 10 Categories After Transformation')

# Rotate and align the tick labels
plt.xticks(rotation=45, ha='right')

# Adjust the subplot layout
plt.tight_layout()

plt.show()
```
